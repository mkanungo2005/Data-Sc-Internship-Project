```python
import pandas as pd
import altair as alt

data = [
    {"category": "Facebook", "group": "Total Installs", "value": 1000000000},
    {"category": "Facebook", "group": "Total Reviews", "value": 78158306},
    {"category": "Whatsapp Messenger", "group": "Total Installs", "value": 1000000000},
    {"category": "Whatsapp Messenger", "group": "Total Reviews", "value": 69119316},
    {"category": "Instagram", "group": "Total Installs", "value": 1000000000},
    {"category": "Instagram", "group": "Total Reviews", "value": 66577313},
    {"category": "Messenger", "group": "Total Installs", "value": 1000000000},
    {"category": "Messenger", "group": "Total Reviews", "value": 56642847},
    {"category": "Clash of Clans", "group": "Total Installs", "value": 100000000},
    {"category": "Clash of Clans", "group": "Total Reviews", "value": 44881447}
]

df = pd.DataFrame(data)

# Pie chart for Total Installs
install_chart = alt.Chart(df[df['group'] == 'Total Installs']).mark_arc(stroke='black',strokeWidth=2).encode(
    theta=alt.Theta(field="value", type="quantitative"),
    color=alt.Color(field="category", type="nominal"),
    tooltip=['category', 'value']
).properties(
    title="Total Installs Distribution"
)

# Pie chart for Total Reviews
review_chart = alt.Chart(df[df['group'] == 'Total Reviews']).mark_arc(stroke='black',strokeWidth=2).encode(
    theta=alt.Theta(field="value", type="quantitative"),
    color=alt.Color(field="category", type="nominal"),
    tooltip=['category', 'value']
).properties(
    title="Total Reviews Distribution"
)

# Display side-by-side
install_chart | review_chart
```

```python
#Total installs vs total reviews for the top 5 apps with the most installs
data = [
    {"category": "Facebook", "group": "Total Installs", "value": 1000000000},
    {"category": "Facebook", "group": "Total Reviews", "value": 78158306},
    {"category": "Whatsapp Messenger", "group": "Total Installs", "value": 1000000000},
    {"category": "Whatsapp Messenger", "group": "Total Reviews", "value": 69119316},
    {"category": "Instagram", "group": "Total Installs", "value": 1000000000},
    {"category": "Instagram", "group": "Total Reviews", "value": 66577313},
    {"category": "Messenger", "group": "Total Installs", "value": 1000000000},
    {"category": "Messenger", "group": "Total Reviews", "value": 56642847},
    {"category": "Clash of Clans", "group": "Total Installs", "value": 100000000},
    {"category": "Clash of Clans", "group": "Total Reviews", "value": 44881447}
]

df = pd.DataFrame(data)

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('category:N', title='App'),
    xOffset='group:N',
    y=alt.Y('value:Q', title='Value'),
    color='group:N'
).properties(
    width=600,
    height=400,
)

chart
```
