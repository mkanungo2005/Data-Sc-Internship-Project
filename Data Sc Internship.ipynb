```python
import pandas as pd
import altair as alt
```

```python
#Line Chart: Average Rating by Category
df = pd.read_csv('Cleaned Dataset.csv')
avg_rating = df.groupby('Category', as_index=False)['Rating'].mean()

alt.Chart(avg_rating).mark_line(point=True).encode(
    x='Category:N',
    y='Rating:Q',
    tooltip=['Category', 'Rating']
).properties(
    width=700,
    title='Average Rating by Category'
)
```

```python
genre_counts = df['Genres'].value_counts().head(10).reset_index()
genre_counts.columns = ['Genre','Count']
#Bar Chart: Top 10 Genres
alt.Chart(genre_counts).mark_bar().encode(
    x=alt.X('Genre:N', sort='-y', title='Genre'),
    y=alt.Y('Count:Q', title='Number of Apps'),
    color='Genre:N',
    tooltip=['Genre', 'Count']
).properties(
    width=700,
    title='Top 10 Genres'
)
```

```python
category_counts = df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']
# Bar Chart: Top Categories
alt.Chart(category_counts.head(10)).mark_bar().encode(
    x=alt.X('Category:N', sort='-y'),
    y='Count:Q',
    color='Category:N',
    tooltip=['Category', 'Count']
).properties(title='Top 10 App Categories')
```

```python
content_rating_counts = df['Content Rating'].value_counts().reset_index()
content_rating_counts.columns = ['Content Rating', 'Count']
# Donut Chart: Content Rating
alt.Chart(content_rating_counts).mark_arc(innerRadius=50).encode(
    theta='Count:Q',
    color='Content Rating:N',
    tooltip=['Content Rating', 'Count']
).properties(title='Content Rating Distribution')
```

```python
type_counts = df['Type'].value_counts().reset_index()
type_counts.columns = ['Type', 'Count']
# Donut Chart: App Type
alt.Chart(type_counts).mark_arc(innerRadius=50).encode(
    theta='Count:Q',
    color='Type:N',
    tooltip=['Type', 'Count']
).properties(title='Free vs Paid Apps')
```

```python
# Pie Chart: App Categories
alt.Chart(category_counts.head(10)).mark_arc().encode(
    theta='Count:Q',
    color='Category:N',
    tooltip=['Category', 'Count']
).properties(title='Top 10 App Categories')
```
